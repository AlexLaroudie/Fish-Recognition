```python
# Définir le chemin du fichier
fichier_entree = 'final_all_index.txt'

fichier_sortie = 'fichier_unique.txt'

# Initialiser un ensemble pour stocker les IDs vus
ids_vus = set()

# Ouvrir le fichier d'entrée et le fichier de sortie
with open(fichier_entree, 'r', encoding='utf-8') as fin, \
     open(fichier_sortie, 'w', encoding='utf-8') as fout:
    
    for ligne in fin:
        # Supprimer les espaces en début et fin de ligne
        ligne = ligne.strip()
        
        # Ignorer les lignes vides
        if not ligne:
            continue
        
        # Séparer l'ID et le reste de la ligne
        parties = ligne.split('=', 1)  # Sépare sur le premier '='
        id_ = parties[0]
        
        # Vérifier si l'ID a déjà été vu
        if id_ not in ids_vus:
            fout.write(ligne + '\n')  # Écrire la ligne dans le fichier de sortie
            ids_vus.add(id_)           # Marquer l'ID comme vu
```

```python
import pandas as pd

def lire_fichier_txt(nom_fichier):
    with open(nom_fichier, 'r', encoding='utf-8') as file:
        lignes = file.readlines()
    
    # Fusionner les lignes si nécessaire en utilisant le point '.' comme délimiteur
    contenu = ''.join(lignes)
    enregistrements = contenu.split('.')
    
    # Nettoyer et séparer les champs par '='
    donnees = []
    for enregistrement in enregistrements:
        # Supprimer les espaces et sauts de ligne
        enregistrement = enregistrement.strip()
        if not enregistrement:
            continue  # Ignorer les enregistrements vides
        # Séparer par '='
        champs = enregistrement.split(',')
        champs = [champ.strip() for champ in champs if champ.strip()]
        donnees.append(champs)
    
    return donnees

def convertir_en_csv(donnees, nom_fichier_csv):
    # Identifier le nombre maximum de colonnes
    max_colonnes = max(len(enregistrement) for enregistrement in donnees)
    
    # Ajouter des en-têtes de colonnes
    colonnes = [f'Champ_{i+1}' for i in range(max_colonnes)]
    
    # Remplir les enregistrements avec des valeurs vides si nécessaire
    donnees_remplies = [enregistrement + ['']*(max_colonnes - len(enregistrement)) for enregistrement in donnees]
    
    # Créer le DataFrame
    df = pd.DataFrame(donnees_remplies, columns=colonnes)
    
    # Sauvegarder en CSV avec des virgules comme séparateurs
    df.to_csv(nom_fichier_csv, index=False, sep=',', encoding='utf-8')
    print(f"Conversion réussie ! Le fichier CSV est enregistré sous le nom '{nom_fichier_csv}'.")

def main():
    nom_fichier_txt = 'final_all_index.txt'  # Remplacez par le nom de votre fichier TXT
    nom_fichier_csv = 'description.csv'  # Nom du fichier CSV de sortie
    
    donnees = lire_fichier_txt(nom_fichier_txt)
    convertir_en_csv(donnees, nom_fichier_csv)

if __name__ == "__main__":
    main()
```
